# cleaning text in description, receiver_address and service columns

```python
!pip install stop_words
# Define function for tokenize and lemmatizing
from stop_words import get_stop_words
from nltk.tokenize import RegexpTokenizer
from nltk.stem.porter import PorterStemmer
```

```python
new_df['description'] = new_df['description'].apply(lambda x: " ".join(x for x in str(x).split() if not x.isdigit() and not x.isspace()))
new_df['description'] = new_df['description'].str.replace('[^\w\s]','')
new_df['description'] = new_df['description'].str.lower()
new_df
```

```python
new_df['receiver_address'] = new_df['receiver_address'].str.upper()
```

```python
new_df['service'] = new_df['service'].str.upper()
```

```python
new_df.info()
```

# change datatype for date, actual_weight, buy_in_store

```python
new_df['date'] = new_df['date'].astype('datetime64[ns]')
new_df['actual_weight'] = new_df['actual_weight'].astype(float)
new_df['buy_in_store'] = new_df['buy_in_store'].astype(float)
```
